**Use these block of codes, If you have thousands of Folders inside of a Folder**

```python
import os
from googleapiclient.discovery import build

# Function to list files in a folder on Google Drive
def list_files(service, folder_id):
    files = []
    page_token = None
    while True:
        response = service.files().list(
            q=f"'{folder_id}' in parents",
            fields="nextPageToken, files(id, name)",
            pageToken=page_token
        ).execute()
        files.extend(response.get('files', []))
        page_token = response.get('nextPageToken', None)
        if page_token is None:
            break
    return files

# Function to download a file from Google Drive
def download_file(service, file_id, file_name):
    request = service.files().get_media(fileId=file_id)
    with open(file_name, 'wb') as file:
        file.write(request.execute())

# Set your Google Drive folder ID here
dataset_folder_id = '1prqLD9Kzrj1XbPP3EO8cLxMOUg1ixmTf'

# Authenticate with Google Drive API
drive_service = build('drive', 'v3')

# Create folders to store Seg and Flair files
os.makedirs('/content/drive/MyDrive/Seg', exist_ok=True)
os.makedirs('/content/drive/MyDrive/Flair', exist_ok=True)

# List files in the dataset folder
files = list_files(drive_service, dataset_folder_id)
i = 0

# Iterate through each folder in the dataset
for folder in files:
    folder_name = folder['name']
    folder_id = folder['id']
    i += 1
    print("The name of Folder: ", i, folder_name)

    # List files in the current folder
    folder_files = list_files(drive_service, folder_id)

    # Iterate through each file in the folder
    for file in folder_files:
        file_id = file['id']
        file_name = file['name']

        # Download _seg.png and _flair.png files
        if '_seg.png' in file_name:
            download_file(drive_service, file_id, f'/content/drive/MyDrive/Seg/{folder_name}_seg.png')
        elif '_flair.png' in file_name:
            download_file(drive_service, file_id, f'/content/drive/MyDrive/Flair/{folder_name}_flair.png')

print("Download completed.")
```
